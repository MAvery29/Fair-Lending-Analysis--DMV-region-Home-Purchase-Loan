```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
DCLoan = pd.read_csv('state_DC.csv')
MDLoan = pd.read_csv('state_MD.csv')
VALoan = pd.read_csv('state_VA.csv')
```

```python
DMVLoan = pd.concat([DCLoan, MDLoan, VALoan], ignore_index=True)
print(DMVLoan.head())
```

```python
print(DMVLoan.head())
```

```python
DMVLoan.isnull().sum()[DMVLoan.isnull().sum() > 0]
```

```python
print(DMVLoan.columns.tolist())
```

```python
ColumnsToKeep = [
	"interest_rate",
	"loan_to_value_ratio",
	"debt_to_income_ratio",
	"property_value",
	"income",
	"applicant_sex",
	"state_code",
	"applicant_credit_score_type",
	"submission_of_application",
    "derived_race",
    "loan_type",
    "action_taken",
    "loan_purpose",
    "loan_term",
    "loan_amount",
    "county_code",
]

DMVLoan[ColumnsToKeep].head()
```

```python
print(DMVLoan[ColumnsToKeep].shape)
print(DMVLoan[ColumnsToKeep].isnull().sum())
```

```python
DMVLoan = DMVLoan[ColumnsToKeep].dropna(subset=["loan_term", "county_code"])

print(DMVLoan.shape)
```

```python
DMVHLoan = DMVLoan[DMVLoan["loan_purpose"] == 1]
print(DMVHLoan.shape)
```

```python
DMVHLoan.value_counts("action_taken")
```

1 = Loan originated (approved and closed)
3 = Application denied
4 = Application withdrawn
6 = Purchased loan

```python
DMVHLoan.value_counts("derived_race")
```

```python
DMVHLoan.value_counts("applicant_sex")
```

1 - Male
2 - Female
3 - Information not provided by applicant in mail, internet, or telephone application
4 - Not applicable
5 - No co-applicant
6 - Co-applicant selected both male and female

```python
DMVHLoan["action_taken_label"] = DMVHLoan["action_taken"].replace({
	1: "Approved",
    2: "Approved but not accepted",
	3: "Denied",
	4: "Withdrawn",
    5: "file closed for incompleteness",
	6: "Loan Purchased",
    7: "Preapproval Denied",
    8: "Preapproval approved but no accepted"
})

DMVHLoan["applicant_sex_label"] = DMVHLoan["applicant_sex"].replace({
	1: "Male",
	2: "Female",
    3: "Information Not Provided",
    4: "Not Applicable",
    6: "Applicant selected both male and female"
})

DMVHLoan[["action_taken_label", "applicant_sex_label"]].head()
```

```python
DMVHLoan["action_taken_label"].value_counts(normalize=True)
```

```python
ApprovalRate = DMVHLoan[DMVHLoan["action_taken_label"].isin(["Approved", "Denied"])]
```

```python
ApprovalByRace = (
    ApprovalRate
    .groupby(["derived_race", "action_taken_label"])
    .size()
    .reset_index(name="count")
    .assign(
        proportion=lambda df: df.groupby("derived_race")["count"].transform(lambda s: s / s.sum())
    )
    [["derived_race", "action_taken_label", "proportion"]]
)

print(ApprovalByRace)
```

```python
ApprovalByRace.to_csv("ApprovalByRace.csv", index=False)
```

```python
ApprovalBySex = (
    ApprovalRate
    .groupby(["applicant_sex_label", "action_taken_label"])
    .size()
    .reset_index(name="count")
    .assign(
        proportion=lambda df: df.groupby("applicant_sex_label")["count"].transform(lambda s: s / s.sum())
    )
    [["applicant_sex_label", "action_taken_label", "proportion"]]
)

print(ApprovalBySex)
```

```python
ApprovalBySex.to_csv("ApprovalBySex.csv", index=False)
```

```python
LoanAmountByRace = (
    DMVHLoan
    .groupby("derived_race")["loan_amount"]
    .mean()
    .reset_index(name="loan_amount")
)
LoanAmountByRace = round(LoanAmountByRace, 2)

print(LoanAmountByRace)
```

```python
ApprovalRateByState = (
    ApprovalRate
    .assign(approved=lambda df: df["action_taken_label"] == "Approved")
    .groupby("state_code")["approved"]
    .mean()
    .reset_index(name="approval_rate")
)

print(ApprovalRateByState)

print(ApprovalRateByState)
```
